```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
#Transpose the DataFrame to assign the borough names as the first entry in each row, then show the first few rows to confirm
properties = properties.transpose()
properties.head()
```

```python
#Reset the DF index so that each borough can be accessed via index starting from 0
properties = properties.reset_index()
```

```python
#Confirm new indices are a range
properties.index
```

```python
#Assign the values of each column in the first row as the column headers, scuh that each year's datetime is the title for each column
#Drop the first row, the values of which are now the column headers
properties.columns=properties.iloc[0]
properties.drop(index=properties.index[0], 
        axis=0, 
        inplace=True)

#Check .head() to confirm the change
properties.head()
```

**2.3. Cleaning the data (part 2)**

You might have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
#View all columns as a list to see what needs renaming
properties.columns
```

```python
#Rename the column containing borough names "Borough"
properties = properties.rename(columns = {'Unnamed: 0': 'Borough', pd.NaT:'ID'})
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
#Melt the columns and rename the label as "Month" and the values as "Average Price"
properties_tall = pd.melt(properties, id_vars= ['Borough', 'ID'], var_name='Month', value_name='Average Price')
```

```python
#Check head again
properties_tall.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
#Return types of all columns
properties_tall.dtypes
```

```python
#Average Price must be converted to a float; coerce the "-" in the 2021 values
properties_tall['Average Price'] = pd.to_numeric(properties_tall['Average Price'], errors='coerce')
```

```python
#Confirm the type change
properties_tall.dtypes
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
#Compare the number of observations in all columns; in theory, they should match (i.e., all boroughs should have an id and a price for all months on record)
properties_tall.count()
```

```python
#View all the Boroughs to see what's going on (use unique() to see one example of all Boroughs in the DF)
properties_tall['Borough'].unique()
```

```python
#Everything after 'Westminster is not a London Borough'; subset an example to see contents
properties_tall[properties_tall['Borough'] == 'Unnamed: 34']
```

```python
#View another example for one of the other anaomalous entries other than an "Unnamed"
properties_tall[properties_tall['Borough'] == 'EAST MIDLANDS']

#Note - East Midlands, Yorskshire, etc. are other parts of England - not London boroughs even though they have valid ID and Average Price values
```

```python
#Use a boolean filter to create a new DF holding only rows that *don't* have a NaN in the ID column
valid_properties = properties_tall[properties_tall['ID'].notna()]
```

```python
#View head to at least row 33 to confirm "Unnamed" have been filtered out
valid_properties.head(35)
```

```python
#Remove all non-Borough observations manually by name
not_boroughs= ['Inner London', 'Outer London', 'NORTH EAST',
       'NORTH WEST', 'YORKS & THE HUMBER', 'EAST MIDLANDS',
       'WEST MIDLANDS', 'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST',
       'SOUTH WEST', 'England']

boroughs_only_properties = valid_properties[~valid_properties['Borough'].isin(not_boroughs)]
```

```python
#Verify again
boroughs_only_properties.head(35)
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
#Prepare by naming final DataFrame 'df' for ease of use
df = boroughs_only_properties
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:
1. look through the `Month` column

```python
#Use lamda function as suggested to isolate date portion of Month datetime and put into new column "Year"
pd.options.mode.chained_assignment = None 

df['Year'] = df['Month'].apply(lambda t: t.year)

df.head()
```

```python
#Aggregate each borough by year and get average (mean) of Avg. Price
grouped_df = df.groupby(by=["Borough", "Year"]).mean()

grouped_df.head(40)
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#Create a new column in the DF containing the ratio of the first price on record divided by the last
#Sort the DF by this column to see the largest ratio

#Step 1 - Iterate over df and find the ratio between the first price on record and the last for each borough; store in a dict

dif_dict = {}

for index, row in df.iterrows():
    current_borough = df.loc[index, "Borough"]
    
    borough_rows = df[df["Borough"] == current_borough]
    
    dif_value = borough_rows.iloc[1]["Average Price"] / borough_rows.iloc[-1]["Average Price"]
    
    dif_dict[current_borough] = dif_value

print(dif_dict)
```

```python
#Step 2 - Create a new DF from the dict so that it can be sorted by the dif_value

dif_df = pd.DataFrame.from_dict(dif_dict, orient='index')

dif_df.columns = ['Avg. Price Ratio']

dif_df.head()
```

```python
#Step 3 - Sort by Avg. Price Ratio; add inplace flag so order is persisted for visualizing
dif_df.sort_values('Avg. Price Ratio', ascending=False, inplace=True)
```

```python
#Visualize the ratio of change for each borrow for easy comparison
dif_df.plot(kind='bar')
plt.show()
```

```python
#Repeat steps above, but this time finding difference between year 1 and year n to see the absoute price increases

dif_dict_abs = {}

for index, row in df.iterrows():
    current_borough = df.loc[index, "Borough"]
    
    borough_rows = df[df["Borough"] == current_borough]
    
    dif_value = borough_rows.iloc[-1]["Average Price"] - borough_rows.iloc[1]["Average Price"]
    
    dif_dict_abs[current_borough] = dif_value

print(dif_dict_abs)
```

```python
dif_df_abs = pd.DataFrame.from_dict(dif_dict_abs, orient='index')

dif_df_abs.columns = ['Avg. Price Ratio']

dif_df_abs.head()
```

```python
dif_df_abs.sort_values('Avg. Price Ratio', ascending=False, inplace=True)
dif_df_abs.head()
```

```python
dif_df_abs.plot(kind="bar")
plt.show()
```
